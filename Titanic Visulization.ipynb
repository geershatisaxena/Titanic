#### Now visualizing the Titatnic dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("titanic")
```

#### Distribution & Histogram Plots

```python
#  Age distribution
sns.histplot(df['age'], bins=30, color='skyblue')
plt.title("Age Distribution")

```

```python
#  Fare distribution
sns.histplot(df['fare'], bins=30, color='green')
plt.title("Fare Distribution")
```

```python
#  Age histogram (Matplotlib)
plt.hist(df['age'].dropna(), bins=20, color='orange')
plt.title("Age Histogram")
```

#### Heatmaps

```python
# Correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
```

```python
#  Survival rate pivot heatmap (class vs sex)
pivot = df.pivot_table(index='class', columns='sex', values='survived')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("Survival Rate by Class and Sex")
```

```python
# Fare by class and sex
fare_pivot = df.pivot_table(index='class', columns='sex', values='fare')
sns.heatmap(fare_pivot, annot=True, cmap='Reds')
plt.title("Fare by Class and Gender")
```

```python
#  Age by class and sex
age_pivot = df.pivot_table(index='class', columns='sex', values='age')
sns.heatmap(age_pivot, annot=True, cmap='Greens')
plt.title("Age by Class and Gender")
```

```python
#  Count by class and who
count = pd.crosstab(df['class'], df['who'])
sns.heatmap(count, annot=True, fmt='d', cmap='Purples')
plt.title("Count of Who by Class")
```

```python
#  Correlation on selected columns
selected = df[['survived', 'age', 'fare', 'sibsp', 'parch']]
sns.heatmap(selected.corr(), annot=True, cmap='RdYlGn')
plt.title("Selected Feature Correlation")
```

```python
# Fare correlation with other features
fare_corr = df.corr(numeric_only=True)['fare'].sort_values(ascending=False)
fare_corr.plot(kind='bar', color='firebrick')
plt.title("Fare Correlation")
plt.show()
```

```python
#  Heatmap with diverging palette
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='RdBu_r')
plt.title("Correlation Heatmap with Diverging Colors")
plt.show()
```

```python
#  Class survival rate (manual calc)
manual = df.groupby('class')['survived'].mean().to_frame().T
sns.heatmap(manual, annot=True, cmap='cividis')
plt.title("Manual Class Survival Rate")
```

```python
# Embark vs survived count heatmap
embark_survive = pd.crosstab(df['embarked'], df['survived'])
sns.heatmap(embark_survive, annot=True, cmap='Greens')
plt.title("Survival by Embark Port")
```
